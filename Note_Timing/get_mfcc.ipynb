```python
import sys
import os
import numpy as np
import pandas as pd
import json
import pickle
from zipfile import ZipFile
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
def get_mfcc(song_file, start_interval=10, end_interval=17):
    y, sr = librosa.load(song_file)
    start_time = librosa.time_to_samples(start_interval)
    end_time = librosa.time_to_samples(end_interval)

    required_slice = y[int(start_time):int(end_time)]
    required_mfcc = librosa.feature.mfcc(y=required_slice, sr=sr)

    # print(required_mfcc)

    return required_mfcc
```

```python
def plot_mfcc(test_mfcc, my_title="Title"):
    fig, ax = plt.subplots()
    img = librosa.display.specshow(test_mfcc, x_axis='time', ax=ax)
    fig.colorbar(img, ax=ax)
    ax.set(title=my_title)
```

```python
plot_mfcc(get_mfcc())
```
